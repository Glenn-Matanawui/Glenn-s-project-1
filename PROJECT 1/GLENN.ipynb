```python
mean_FA = data_set.fixed_acidity.mean()
mean_VA = data_set.volatile_acidity.mean()
mean_CA = data_set.citric_acid.mean()
mean_RS = data_set.residual_sugar.mean()
mean_C = data_set.chlorides.mean()
mean_FSD = data_set.free_sulfur_dioxide.mean()
mean_TSD = data_set.total_sulfur_dioxide.mean()
mean_D = data_set.density.mean()
mean_pH = data_set.pH.mean()
mean_S = data_set.sulphates.mean()
mean_A = data_set.alcohol.mean()
mean_Q = data_set.quality.mean()
```

```python
print(f'fixed_acidity: {mean_FA}')
print(f'volatile_acidity: {mean_VA}')
print(f'citric_acid: {mean_CA}')
print(f'residual_sugar: {mean_RS}')
print(f'chlorides: {mean_C}')
print(f'free_sulfur_dioxide: {mean_FSD}')
print(f'total_sulfur_dioxide: {mean_TSD}')
print(f'density: {mean_D}')
print(f'pH: {mean_pH}')
print(f'sulphates: {mean_S}')
print(f'alcohol: {mean_A}')
print(f'quality: {mean_Q}')
```

```python
data_set['fixed_acidity'] = data_set['fixed_acidity'].fillna(mean_FA)
data_set['volatile_acidity'] = data_set['volatile_acidity'].fillna(mean_VA)
data_set['citric_acid'] = data_set['citric_acid'].fillna(mean_CA)
data_set['residual_sugar'] = data_set['residual_sugar'].fillna(mean_RS)
data_set['chlorides'] = data_set['chlorides'].fillna(mean_C)
data_set['free_sulfur_dioxide'] = data_set['free_sulfur_dioxide'].fillna(mean_FSD)
data_set['pH'] = data_set['pH'].fillna(mean_pH)
data_set['sulphates'] = data_set['sulphates'].fillna(mean_S)
data_set['alcohol'] = data_set['alcohol'].fillna(mean_A)
data_set['quality'] = data_set['quality'].fillna(mean_Q)
```

# This is a Heatmap

```python
import matplotlib.pyplot as plt
plt.imshow(data_set.corr(), cmap='coolwarm', interpolation='none')
plt.colorbar()
plt.xticks(range(len(data_set.columns)), data_set.columns, rotation=90)
plt.yticks(range(len(data_set.columns)), data_set.columns)
plt.gcf().set_size_inches(14,12)

labels = data_set.corr().values
for y in range(labels.shape[0]):
    for x in range(labels.shape[1]):
        plt.text(x, y, '{:.2f}'.format(labels[y, x]), ha='center', va='center', color='white')
```
